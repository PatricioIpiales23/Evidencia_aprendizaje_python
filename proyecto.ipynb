```python
import pandas as pd
```

```python
print('Este programa permite conocer las cantidades exactas de reactivos' + \
      ' para preparar soluciones químicas.\n')
print('Lista de reactivos comúnmente utilizados:')

base_datos_df = pd.read_excel('Reactivos y solventes_base de datos.xlsx')
base_datos_df.index.name = 'Número'
base_datos_df.drop(labels=0, inplace=True)
base_datos_rename_df=base_datos_df.rename(columns={
    'Unnamed: 0': 'Reactivo',
    'Unnamed: 1': 'Fórmula reactivo',
    'Unnamed: 2': 'Peso molecular reactivo (g/mol)',
    'Unnamed: 5': 'Tipo reactivo',
    'Unnamed: 6': 'Peso equivalente',
    'Unnamed: 8': 'Solvente',
    'Unnamed: 9': 'Fórmula solvente',
    'Unnamed: 10':'Estado físico',
    'Unnamed: 11': 'Densidad (g/mL)',
    'Unnamed: 12': 'Peso molecular solvente (g/mol))'
})

base_datos_rename_df[['Reactivo', 'Fórmula reactivo', 'Tipo reactivo']]
```

```python
print('\nSolventes comúnmente utilizados:')
base_datos_rename_df.head(7)[['Solvente','Fórmula solvente', 'Estado físico']]
```

```python
expresion_concentraciones = pd.Series(['masa-masa (% p/p)', 'masa-volumen (% m/v)',
                                       'volumen-volumen (% v/v)','Molaridad (M)',
                                       'Molalidad (m)', 'Normalidad (N)'],
                                       index=[1, 2, 3, 4, 5, 6], name='Concentración')

print('Formas de expresar la concentración de la solución:')
expresion_concentraciones
```

*Entradas generadas por el usuario*

```python
print('Por favor, ingrese la información que se solicita a continuación en base a las ' + \
      'tablas anteriormente desplegadas.\nEl programa no seguirá su curso normal mientras ' +\
      'no ingrese correctamente la información solicitada.\n')

# Selección del reactivo
reactivo = int(input('Ingrese el número del reactivo: '))
while not 1 <= reactivo <= 28:
    reactivo = int(input('Ingrese el número del reactivo: '))

# Selección del solvente
solvente = int(input('Ingrese el número del solvente: '))
while not 1 <= solvente <= 7:
    solvente = int(input('Ingrese el número del reactivo: '))

# Selección del volumen de solución
volumen_solucion = float(input('Ingrese el volumen de solución que desea preparar (mL):'))
while volumen_solucion < 0:
    volumen_solucion = float(input('Ingrese el volumen de solución que desea preparar (mL):'))

# Selección del tipo de concentración para preparar
tipo_concentracion = int(input('Ingrese el número del tipo de concentración que desea preparar:'))
while not 1 <= tipo_concentracion <= 6:
    tipo_concentracion = int(input('Ingrese el número del tipo de concentración que desea preparar:'))

if (reactivo == base_datos_rename_df.index[12] or reactivo == base_datos_rename_df.index[14]
                                     or reactivo == base_datos_rename_df.index[22]):
    tipo_concentracion = int(input('El reactivo seleccionado no posee un peso equivalente ' + \
                        'para calcular normalidad. Por favor, ingrese otro ' + \
                        'número del tipo de concentración que desea preparar:'))
    while tipo_concentracion == 6:
        tipo_concentracion = int(input('El reactivo seleccionado no posee un peso equivalente ' + \
                        'para calcular normalidad. Por favor, ingrese otro ' + \
                        'número del tipo de concentración que desea preparar:'))

# Valor numérico de la concentración
valor_concentracion = float(input('Ingrese el valor numérico de la concentración que desea preparar:'))
while valor_concentracion < 0:
    valor_concentracion = float(input('Ingrese el valor numérico de la concentración que desea preparar:'))

# Datos adicionales
pureza_reactivo = float(input('Ingrese la pureza del reactivo que está a su disposición (%p/p):'))
densidad_reactivo = float(input('Ingrese la densidad (gravedad especifica) del reactivo que está a su '+ \
                       'disposición (g/mL):'))
PM = float(base_datos_rename_df['Peso molecular reactivo (g/mol)'][reactivo])   # peso molecular 
# del reactivo [g/mol]

volumen_solucion_litros = volumen_solucion / 1000
densidad_solvente = base_datos_rename_df['Densidad (g/mL)'][solvente]
masa_solvente = volumen_solucion * densidad_solvente

# Calculos 
if tipo_concentracion == 1:
    masa_reactivo_puro = ((valor_concentracion / 100) * masa_solvente) / (1 - valor_concentracion /100)
    masa_reactivo_impuro = masa_reactivo_puro * (1 / (pureza_reactivo / 100))
    print(f'RESULTADO\nPara preparar una solución de {base_datos_rename_df['Fórmula reactivo'][reactivo]}' + \
        f' en {volumen_solucion} mL de {base_datos_rename_df['Fórmula solvente'][solvente]} ' + \
        f'al {valor_concentracion} de {expresion_concentraciones.loc[tipo_concentracion]}, se debe' + \
        f' pesar\n{masa_reactivo_impuro:.2f} g de reactivol al {pureza_reactivo} (%p/p) de pureza.')
    
elif tipo_concentracion == 2:
    masa_reactivo_puro = (valor_concentracion / 100) * volumen_solucion
    masa_reactivo_impuro = masa_reactivo_puro *(1 / (pureza_reactivo / 100))
    print(f'RESULTADO\nPara preparar {volumen_solucion} mL de solución de ' + \
          f'{base_datos_rename_df['Fórmula reactivo'][reactivo]}' + \
          f' en {base_datos_rename_df['Fórmula solvente'][solvente]} al {valor_concentracion}' +\
          f' de {expresion_concentraciones.loc[tipo_concentracion]}, se debe pesar\n' + \
          f'{masa_reactivo_impuro:.2f} g del reactivo al {pureza_reactivo} (%p/p) de pureza.')
    
elif tipo_concentracion == 3:
    masa_reactivo_puro = (valor_concentracion / 100) * volumen_solucion * densidad_reactivo
    masa_reactivo_impuro = masa_reactivo_puro * (1 / (pureza_reactivo / 100))
    print(f'RESULTADO\nPara preparar {volumen_solucion} mL de solución de ' + \
          f'{base_datos_rename_df['Fórmula reactivo'][reactivo]}' + \
          f' en {base_datos_rename_df['Fórmula solvente'][solvente]} al {valor_concentracion}' +\
          f' de {expresion_concentraciones.loc[tipo_concentracion]}, se debe pesar\n' + \
          f'{masa_reactivo_impuro:.2f} g del reactivo al {pureza_reactivo} (%p/p) de pureza.')
    
elif tipo_concentracion == 4:
    masa_reactivo_puro = valor_concentracion * volumen_solucion_litros * PM
    masa_reactivo_impuro = masa_reactivo_puro * (1 / (pureza_reactivo / 100))
    print(f'RESULTADO\nPara preparar {volumen_solucion} mL de solución de ' + \
          f'{base_datos_rename_df['Fórmula reactivo'][reactivo]}' + \
          f' en {base_datos_rename_df['Fórmula solvente'][solvente]} al {valor_concentracion}' +\
          f' de {expresion_concentraciones.loc[tipo_concentracion]}, se debe pesar\n' + \
          f'{masa_reactivo_impuro:.2f} g del reactivo al {pureza_reactivo} (%p/p) de pureza.')
    
elif tipo_concentracion == 5:
    masa_reactivo_puro = ((valor_concentracion * volumen_solucion * densidad_solvente) / 1000) * PM
    masa_reactivo_impuro = masa_reactivo_puro * (1 / (pureza_reactivo / 100))
    print(f'RESULTADO\nPara preparar una solución de {base_datos_rename_df['Fórmula reactivo'][reactivo]}' + \
        f' en {volumen_solucion} mL de {base_datos_rename_df['Fórmula solvente'][solvente]} ' + \
        f'al {valor_concentracion} de {expresion_concentraciones.loc[tipo_concentracion]}, se debe pesar\n' + \
        f'{masa_reactivo_impuro:.2f} g del reactivo al {pureza_reactivo} (%p/p) de pureza.')
    
elif tipo_concentracion == 6:
    masa_reactivo_puro = (volumen_solucion_litros * valor_concentracion *
                          ( 1 / base_datos_rename_df['Peso equivalente'][reactivo]) * PM)
    masa_reactivo_impuro = masa_reactivo_puro * (1 / (pureza_reactivo / 100))
    print(f'RESULTADO\nPara preparar una solución de {volumen_solucion} mL de ' + \
        f'{base_datos_rename_df['Fórmula reactivo'][reactivo]} en ' + \
        f'{base_datos_rename_df['Fórmula solvente'][solvente]} ' + \
        f'al {valor_concentracion} de {expresion_concentraciones.loc[tipo_concentracion]}, se debe pesar\n' + \
        f'{masa_reactivo_impuro:.2f} g del reactivo al {pureza_reactivo} (%p/p) de pureza.')
```
